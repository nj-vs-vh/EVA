# Data visualization

```python
from matplotlib import pyplot as plt

from cr_knee_fit import experiments
from cr_knee_fit.elements import Element
from cr_knee_fit.fit_data import Data, DataConfig
```

# All available data

```python
ALL = experiments.DIRECT + experiments.INDIRECT
conf = DataConfig(
    experiments_elements=list(ALL),
    experiments_all_particle=ALL,
    experiments_lnA=ALL,
    elements=Element.regular(),
    elements_R_bounds=(5e2, 1e8),
)

d = Data.load(conf, verbose=True)
fig, ax = plt.subplots(nrows=2, figsize=(10, 15))
fig = d.plot(scale=2.8, figure=fig)
```

# Latest data from DAMPE and LHAASO

```python
conf = DataConfig(experiments_elements=[experiments.dampe], elements=[Element.O, Element.Fe])

d = Data.load(conf, verbose=True)
fig = d.plot(scale=2.8)
```
