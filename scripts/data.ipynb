# Data visualization

```python
from matplotlib import pyplot as plt

from cr_knee_fit import experiments
from cr_knee_fit.elements import Element
from cr_knee_fit.fit_data import Data, DataConfig
```

# All available data

```python
ALL = experiments.DIRECT + experiments.INDIRECT
conf = DataConfig(
    # experiments_elements=list(ALL),
    # experiments_all_particle=ALL,
    experiments_lnA=ALL,
    default_elements=Element.regular(),
    experiments_elements=[
        experiments.dampe,
        experiments.lhaaso_epos,
        experiments.ams02,
        experiments.calet,
    ],
    experiments_all_particle=[
        experiments.hawc,
        experiments.lhaaso_epos,
        experiments.kascade_grande_sibyll,
    ],
    # experiments_lnA=[],
    # default_elements=[
    #     Element.H,
    #     Element.He,
    # ],
    elements_R_bounds=(5e2, 1e10),
)

d = Data.load(conf, verbose=True)
fig, ax = plt.subplots(nrows=2, figsize=(10, 15))
fig = d.plot(scale=2.8, figure=fig)
```
