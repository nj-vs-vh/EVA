# Data visualization

# All available data

```python
ALL = experiments.DIRECT + experiments.INDIRECT
conf = DataConfig(
    # experiments_elements=list(ALL),
    # experiments_all_particle=ALL,
    # experiments_lnA=ALL,
    experiments_elements=[experiments.ice_top_sibyll, experiments.lhaaso_sibyll],
    experiments_all_particle=[experiments.lhaaso_sibyll, experiments.kascade_sibyll],
    experiments_lnA=[],
    # elements=Element.regular(),
    default_elements=[
        Element.H,
        Element.He,
    ],
    elements_R_bounds=(5e2, 1e10),
)

d = Data.load(conf, verbose=True)
fig, ax = plt.subplots(nrows=2, figsize=(10, 15))
fig = d.plot(scale=2.8, figure=fig)
```
