```python
import numpy as np
```

```python
from matplotlib import pyplot as plt

nrows = 6
fig, axes = plt.subplots(nrows=nrows, sharex=True, sharey=True, figsize=(8, 8))

n_pts = 5
x = np.linspace(0, 10, n_pts)

err_scale = 1.0

for ax, corr in zip(axes, np.linspace(0, 1, nrows)):
    for _ in range(10):
        cov = err_scale**2 * (np.eye(n_pts) + corr * (np.ones((n_pts, n_pts)) - np.eye(n_pts)))
        err = np.random.multivariate_normal(mean=np.zeros(n_pts), cov=cov)
        ax.errorbar(x, err, err_scale, fmt=".-")
        ax.set_xticks([])
        ax.set_yticks([])
        ax.text(
            0.01,
            0.05,
            f"$ \\rho = {corr:.2f} $",
            horizontalalignment="left",
            transform=ax.transAxes,
        )
```
